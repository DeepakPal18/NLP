### Mini project - extracting data from NLTK package

```python
from nltk.corpus import movie_reviews
movie_reviews.fileids()
```

```python
# taking any one txt file - 
movie_reviews.words('neg/cv013_10494.txt')
```

```python
# joining the list elements with space in between like sentences-
complete_sen = " ".join(movie_reviews.words('neg/cv013_10494.txt'))
complete_sen
```

```python
# sentence tokenize - 
sent_tokenize(complete_sen)
```

```python
# word tokenize - 
word_tokenize(complete_sen)
```

```python
# whitespace tokenize - comma,fullstops removed
ws_tokenize.tokenize(complete_sen)
```

```python
word_punc = nltk.tokenize.WordPunctTokenizer()
word_punc.tokenize(complete_sen)
```

```python
# stop words - 
stop_words = set(stopwords.words('english'))
sent_filter = [w for w in complete_sen if not w in stop_words]
sent_filter
# this is giving alphabets and no word-
```

```python
# to visaulize while applying stopwords
#!pip install wordcloud
from wordcloud import WordCloud,STOPWORDS
```

```python
word_stop = nltk.corpus.stopwords.words('english')
word_stop
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,16))
wc = WordCloud(max_words=50,stopwords=word_stop)
wc.generate(complete_sen)
plt.imshow(wc)
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,16))
wc = WordCloud(max_words=50,stopwords=word_stop+ ['like','try','even']) # remove some words
wc.generate(complete_sen)
plt.imshow(wc)
plt.show()
```
